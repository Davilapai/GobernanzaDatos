```python
#Librería requeridas
import pandas as pd # Librería para manejo y analisis de estructura de datos
import numpy as np # Librería para matrices y arrays multidimensionales de gran tamaño
import zipfile # Libreria para trabajar con formatos de extensión *ZIP
import os # Librería que proporciona una interfaz para interactuar con el sistema operativo
import datetime # Librería que proporciona funciones para trabajar con fechas y horas
```

#Consultar el archivo ZIP cargado en el entorno web

```python
if zipfile.is_zipfile("Datos_MedidoresAgua.zip"):
    with zipfile.ZipFile("Datos_MedidoresAgua.zip", "r") as archive:
        archive.printdir()
else:
    print("File is not a zip file")
```

```python
#Extraer información de un solo conjunto de datos
with zipfile.ZipFile("Datos_MedidoresAgua.zip", mode="r") as archive:
    info = archive.getinfo("Datos_MedidoresAgua/ConjuntoDatos_MedidoresAgua/ConjuntoDatos_1_50.txt")
```

```python
info.file_size #Información del conjunto de datos en tamaño en bytes del archivo original
```

```python
info.compress_size #Información del conjunto de datos en tamaño en bytes del archivo comprimido
```

```python
info.filename #Infomración del nombre del conjunto de datos
```

```python
info.date_time #Información de modificación del archivo
```

```python
#Extraer información de todos los conjuntos de datos en el archivo *ZIP
with zipfile.ZipFile("Datos_MedidoresAgua.zip", mode="r") as archive:
    for info in archive.infolist():
        print(f"Filename: {info.filename}")
        print(f"Modified: {datetime.datetime(*info.date_time)}")
        print(f"Normal size: {info.file_size} bytes")
        print(f"Compressed size: {info.compress_size} bytes")
        print("-" * 20)
```

```python
#Lectura rápida con secuencia bytes "b" de un conjunto de datos del archivo sin extraerlo
with zipfile.ZipFile("Datos_MedidoresAgua.zip", mode="r") as archive:
    for line in archive.read("Datos_MedidoresAgua/ConjuntoDatos_MedidoresAgua/ConjuntoDatos_1_50.txt").split(b"\n"):
        print(line)
```

```python
# Extraccion del formato *xslx del archivo *ZIP y descompresión
with zipfile.ZipFile("Datos_MedidoresAgua.zip", mode="r") as archive:
    for file in archive.namelist():
      if file.endswith(".xlsx"):
        archive.extract(file, "/content/Diccionarios")
```

```python
# Extraccion del formato *txt del archivo *ZIP y descompresión
with zipfile.ZipFile("Datos_MedidoresAgua.zip", mode="r") as archive:
    for file in archive.namelist():
      if file.endswith(".txt"):
        archive.extract(file, "/content/ConjuntosDatos")
```

#Tranformación de un formato *TXT a dataframe y exportación a formato *XLSX


```python
# Ruta del archivo .txt
file_path = "/content/ConjuntosDatos/Datos_MedidoresAgua/ConjuntoDatos_MedidoresAgua/ConjuntoDatos_1_50.txt"
# Leer el archivo .txt en un DataFrame
# Cambia 'sep' según el delimitador del archivo ','
df = pd.read_csv(file_path, sep=",")  # Cambia ',' por '\t' si es tabulado

# Mostrar las primeras filas del DataFrame
print("Contenido del DataFrame:")
print(df)

#Guardar el DataFrame en otro formato Excel
df.to_excel("CD_1_50.xlsx", index=False)
print("\nEl DataFrame se ha guardado como 'ConjuntoDatos_1_50.xlsx'.")
```

```python
# Ruta del archivo .txt
file_path = "/content/ConjuntosDatos/Datos_MedidoresAgua/ConjuntoDatos_MedidoresAgua/ConjuntoDatos_2_100.txt"
# Leer el archivo .txt en un DataFrame
# Cambia 'sep' según el delimitador del archivo ','
df2 = pd.read_csv(file_path, sep=",")  # Cambia ',' por '\t' si es tabulado

# Mostrar las primeras filas del DataFrame
print("Contenido del DataFrame:")
print(df2)

#Guardar el DataFrame en otro formato Excel
df2.to_excel("CD_2_100.xlsx", index=False)
print("\nEl DataFrame se ha guardado como 'ConjuntoDatos_2_100.xlsx'.")
```
