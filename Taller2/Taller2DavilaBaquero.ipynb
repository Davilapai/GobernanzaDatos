# Tranformación de un formato *TXT a dataframe y exportación a formato *XLSX


Aquí decidimos exportar todos los archivos de una vez, ya que sabemos exactamente el formato de ellos.

```python
# Ruta del archivo .txt

# Carpeta donde están los archivos .txt
ruta_base = "Datos_MedidoresAgua/ConjuntoDatos_MedidoresAgua"

# Carpeta donde guardaremos los archivos .xlsx
ruta_salida = "Datos_MedidoresAgua/DatosXLSX"

# Crear la carpeta de salida si no existe
os.makedirs(ruta_salida, exist_ok=True)

# Crear lista de archivos .txt dentro de la carpeta base
archivos_txt = [f for f in os.listdir(ruta_base) if f.endswith(".txt")]

# Mostrar qué archivos se encontraron
print("Archivos encontrados:")
for archivo in archivos_txt:
    print(" -", archivo)

# Recorrer cada archivo y convertirlo a Excel
for archivo in archivos_txt:
    ruta_archivo = os.path.join(ruta_base, archivo)
    nombre_salida = os.path.splitext(archivo)[0] + ".xlsx"
    ruta_excel = os.path.join(ruta_salida, nombre_salida)

    # Verificamos si ya existe el Excel para no repetir trabajo
    if os.path.exists(ruta_excel):
        print(f"'{nombre_salida}' ya existe en la carpeta de salida. Omitido.")
        continue

    # Leer y convertir a Excel
    df = pd.read_csv(ruta_archivo, sep=",")
    df.to_excel(ruta_excel, index=False)

    print(f"Convertido: {archivo} → {ruta_excel}")

print("\nConversión completada con éxito. Archivos guardados en:", ruta_salida)
```

# Unificar los datasets

```python
# Carpeta donde están los .xlsx individuales
ruta_salida = "Datos_MedidoresAgua/DatosXLSX"

# Listar todos los archivos .xlsx
archivos_xlsx = [f for f in os.listdir(ruta_salida) if f.endswith(".xlsx") and not f.startswith("CD_unificado")]

# Función para extraer el número del nombre (por ejemplo, 10 de 'ConjuntoDatos_10_500.xlsx')
def numero_archivo(nombre):
    match = re.search(r"ConjuntoDatos_(\d+)_", nombre)
    return int(match.group(1)) if match else float('inf')

# Ordenar por el número del archivo
archivos_xlsx.sort(key=numero_archivo)
print(archivos_xlsx)

ruta_base = os.path.join(ruta_salida, archivos_xlsx[0])
df_unificado = pd.read_excel(ruta_base, header=None)

# Unir los demás archivos
for archivo in archivos_xlsx[1:]:
    ruta_actual = os.path.join(ruta_salida, archivo)
    df2 = pd.read_excel(ruta_actual, header=None)

    # Verificar si la primera fila parece encabezado (más de 70% texto)
    if df2.iloc[0].apply(lambda x: isinstance(x, str)).mean() > 0.7:
        df2 = df2.iloc[1:]

    df2 = df2.reset_index(drop=True)
    df_unificado = pd.concat([df_unificado, df2], ignore_index=True, sort=False)

print("¡¡ Unión exitosa !!")
print(df_unificado.head())  # Mostrar primeras filas

# Guardar el DataFrame unificado
ruta_guardado = os.path.join(ruta_salida, "CD_unificado_500.xlsx")
df_unificado.to_excel(ruta_guardado, index=False, header=False)

print(f"\nEl DataFrame se ha guardado como '{ruta_guardado}'.")
```

# Curación de los datos 

### Analisis de errores básico

```python
df = pd.read_excel("Datos_MedidoresAgua/DatosXLSX\CD_unificado_500.xlsx")

#Vamos a analizar primero las falencias y generalidades

print("Dimensiones:", df.shape)
print("Tipos de datos:")
print(df.dtypes)
print("Valores faltantes:")
print(df.isnull().sum())

print("Valores únicos por columna:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} únicos")

print(df.head())
```

### Valores nulos

```python
df = df.dropna(axis=1, thresh=len(df) * 0.8)  # elimina columnas con más de 80% nulo
```

### Duplicados

```python
print(df.duplicated().sum()) #Como no hay, puedo seguir con el resto.
```

### Corregir tipos de datos

```python
cols_fechas = ['Reading Date', 'Previous Reading Date']
df[cols_fechas] = df[cols_fechas].apply(lambda x: pd.to_datetime(x, errors = 'coerce'))

cols_categoricas = ['Reading Frequency', 'Reader ID', 'Type of Contract', 'Reading Validity', 'Certification on the ERP', 'Reason for Reading']
df[cols_categoricas] = df[cols_categoricas].apply(lambda x: x.astype('category'))
print("Datos identificados !")
```

### Validez de los datos

Hay que verificar valores negativos y con valores sin sentido, como superiores a 1000


```python
def eliminar_outliers_iqr(df, columnas):
  for col in columnas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR

    df = df[(df[col] >= limite_inf) & (df[col] <= limite_sup) & (df[col] >= 0)]
  return df
#Este es el método de Tukey
#Es usado en curación de datos

columnas_enteras = ['Reading Value', 'Final Billing']
df = eliminar_outliers_iqr(df, columnas_enteras)
```

### Normalización de los datos


```python
import unicodedata #Para quitar acentos como á é í ó ú

#Columnas de texto:
cols_texto = ['Water Meter ID', 'Reading ID', 'Reading Frequency', 'Reader ID', 'Type of Contract', 'Reading Validity', 'Reason for Reading']


def limpiar_texto(x):
    if isinstance(x, str):
        x = x.lower() #tolowercase
        x = x.strip() #espacios extras
        # quitar acentos (vi el método en internet)
        x = ''.join(
            c for c in unicodedata.normalize('NFKD', x)
            if not unicodedata.combining(c)
        )
        # eliminar comillas y caracteres raros
        x = x.replace('"', '').replace("", "")
        x = ' '.join(x.split()) #Espacios dobles
    return x

df[cols_texto] = df[cols_texto].applymap(limpiar_texto)
```

### Agrupar por categorías

```python
cols_categoricas = [
    'Reading Frequency',
    'Type of Contract',
    'Reading Validity',
    'Reason for Reading',
    'Accessibility'
    'Certifica'
]

for col in cols_categoricas:
    if col in df.columns:
        print(f"\nColumna: {col}")
        print(df[col].value_counts(dropna=False))
```

### -> La siguiente parte me ayudó una IA, normalizamos los lenguajes español, italiano e inglés a español

```python
# Diccionario de reemplazos

reemplazos = {
    'Reading Frequency': {
        'mensual': 'mensual', 'mensile': 'mensual',
        'monthly': 'mensual',
        'bimonthly': 'bimestral', 'bimestral': 'bimestral', 'bimestrale': 'bimestral'
    },
    'Type of Contract': {
        'residential': 'residencial', 'residencial': 'residencial', 'residenziale': 'residencial', 'residental': 'residencial',
        'commercial': 'comercial', 'comercial': 'commercial', 'comerciale': 'commercial',
        'industrial': 'industrial', 'industriale': 'industrial'
    },
    'Reading Validity': {
        'valid': 'valido', 'valido': 'valido',
        'invalid': 'invalido', 'invalido': 'invalido', 'non valido': 'invalido',
        'suspicious': 'sospechoso', 'sospechoso': 'sospechoso', 'sospetto': 'sospechoso'
    },
    'Reason for Reading': {
        'routine': 'rutina', 'rutina': 'rutina', 'routina': 'rutina',
        'first reading': 'primera lectura', 'primera lectura': 'primera lectura', 'prima lettura': 'primera lectura',
        'property vacant': 'propiedad vacante', 'propiedad vacante': 'propiedad vacante', 'immobile vuoto': 'propiedad vacante',
        'zero consumption': 'consumo cero', 'consumo cero': 'consumo cero', 'consumo zero': 'consumo cero',
        'meter failure': 'falla medidor', 'falla medidor': 'falla medidor', 'guasto contatore': 'falla medidor',
        'estimated': 'estimado', 'estimado': 'estimado', 'stimato': 'estimado'
    }
}

# Aplicar reemplazos
for col, mapping in reemplazos.items():
    if col in df.columns:
        df[col] = df[col].replace(mapping)

# Agrupar valores que no están en el diccionario como "other"
for col in reemplazos.keys():
    if col in df.columns:
        df[col] = df[col].where(df[col].isin(reemplazos[col].values()), other='other')

# Convertir a categoría
for col in reemplazos.keys():
    if col in df.columns:
        df[col] = df[col].astype('category')

print("Categorías normalizadas correctamente.")
```
